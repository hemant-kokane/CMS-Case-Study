## 2. Data Merging

```python
# Changing column name "ï»¿PROVNUM" to "CMS Certification Number (CCN)" from PBJ Dataset

pbj_dns.rename(columns={"ï»¿PROVNUM" : "CMS Certification Number (CCN)"}, inplace=True)
```

```python
pbj_dns.columns[0]
```

```python
# Required column from Providers Dataset

provider_data= provider_data[["CMS Certification Number (CCN)", "Ownership Type", "Number of Certified Beds", "Overall Rating"]]
provider_data.head()
```

```python
provider_data.shape
```

```python
# Required columns from region dataset 

region = region[["CMS Certification Number (CCN)", "CMS Region"]]
region.head()
```

```python
region.shape
```

```python
# Merge the PBJ Dataset with Providers Information Dataset

data = pd.merge(left=pbj_dns, right=provider_data, how="left", left_on="CMS Certification Number (CCN)", right_on="CMS Certification Number (CCN)")
data.head()
```

```python
data.shape
```

```python
# Dropping duplicates so we can merge to final data

region = region.drop_duplicates(subset="CMS Certification Number (CCN)", keep="first")
region.head()
```

```python
region.shape
```

```python
# Data after merging region

final_data = pd.merge(left=data, right=region, how="left", left_on="CMS Certification Number (CCN)", right_on="CMS Certification Number (CCN)")
final_data.head()
```

```python
# Information of final data

final_data.info()
```

```python
final_data.shape
```

## 3. Data Understanding and Preparation

```python
# Changing data type of region from float to int

final_data["CMS Region"] = final_data["CMS Region"].astype("Int64")
```

```python
final_data.head()
```

```python
final_data.shape
```

```python
# Craete region dictionary to map their names

region_mapping = {
    1: "Boston",
    2: "New York:",
    3: "Philadelphia",
    4: "Atlanta",
    5: "Chicago",
    6: "Dallas",
    7: "Kansas City",
    8: "Denver",
    9: "San Francisco",
    10: "Seattle"
}
```

```python
# Change region code to Region name

final_data["Region"] = final_data["CMS Region"].map(region_mapping)
```

```python
final_data.head()
```

```python
final_data.Region.value_counts()
```

```python
# Drop CMS Region Column

final_data.drop("CMS Region", axis = 1, inplace=True)
```

```python
final_data.head()
```

```python
col_name = final_data.columns
```

```python
# Total Hrs Column names

Total_Hrs_col = [col for col in col_name if col.startswith('Hrs') and not (col.endswith('emp') or col.endswith('ctr'))]
Total_Hrs_col
```

```python
# Emp Hrs Column names

emp_Hrs_col = [col for col in col_name if col.endswith('emp')]
emp_Hrs_col
```

```python
# ctr Hrs Column names

ctr_Hrs_col = [col for col in col_name if col.endswith('ctr')]
ctr_Hrs_col
```

```python
# create a new column with Sum of total hrs 

final_data["Total_Hrs"] = final_data[Total_Hrs_col].sum(axis=1)
```

```python
final_data.head()
```

```python
# create a new column with Sum of Emp hrs 

final_data["Emp_Hrs"] = final_data[emp_Hrs_col].sum(axis=1)
final_data.head()
```

```python
# create a new column with Sum of Ctr hrs 

final_data["Ctr_Hrs"] = final_data[ctr_Hrs_col].sum(axis=1)
final_data.head()
```

```python
final_data.shape
```
